```python
sales,menu,members
```

```python
sales_menu= pd.merge(left = sales,right = menu,on = 'product_id', how = 'left')
sales_menu
```

```python
sales_members = pd.merge(sales_menu,members,on='customer_id',how = 'left')
sales_members
```

* ## **1. What is the total amount each customer spent at the restaurant?**

```python
pd.pivot_table(sales_menu,
               values= 'price',
               index = 'customer_id',
               aggfunc = 'sum'
)
```

## **2. How many days has each customer visited the restaurant?**

```python
sales.groupby('customer_id').agg(total_days_visited=('order_date','nunique'))
```

* ## **3. What was the first item from the menu purchased by each customer?**


```python
first_item=sales_menu.groupby('customer_id')[['product_name','order_date']].min()
first_item
```

## **4. What is the most purchased item on the menu and how many times was it purchased by all customers?**

```python
count = sales['product_id'].value_counts().reset_index()
count.columns = ['product_id','count']
most_purchased = pd.merge(count,menu,on = 'product_id',how = 'left').iloc[0][['product_name','count']]
most_purchased
```

## **5. Which item was the most popular for each customer?**

```python
count = sales.groupby(['customer_id','product_id']).size().reset_index(name='count')
most_popular = count.sort_values(['customer_id','count'],ascending=[True,False]).groupby('customer_id').first().reset_index()
most_popular = pd.merge(most_popular,menu,on='product_id',how='left')
most_popular
```

## **6. Which item was purchased first by the customer after they became a member?**

```python
member = sales_members[sales_members['order_date']>=sales_members['join_date']]
first_buy= member.sort_values(['customer_id','order_date']).groupby('customer_id').first()
first_buy
```

## **7. Which item was purchased just before the customer became a member?**

```python
#7.Which item was purchased just before the customer became a member?
member = sales_members[sales_members['order_date']<sales_members['join_date']]
beforemember =member.sort_values(['customer_id','order_date']).groupby('customer_id').last()[['product_name','order_date']]
beforemember
```

## **8. What is the total items and amount spent for each member before they became a member?**

```python
pre_membership = member.groupby('customer_id').agg(
    total_items=('product_id','count'),
    total_amount = ('price','sum'))
pre_membership
```

## **9. If each $1 spent equates to 10 points and sushi has a 2x points multiplier - how many points would each customer have?**

```python
sales_menu['points']=sales_menu.apply(
    lambda x: x['price']*10*2 if x['product_name']=='sushi' else x['price']*10,axis=1
)
total_points =sales_menu.groupby('customer_id').agg(total_points=('points','sum'))
total_points
```

## **10. In the first week after a customer joins the program (including their join date) they earn 2x points on all items, not just sushi - how many points do customer A and B have at the end of January?**

```python
def tot_points(row):
  if pd.notna(row['join_date']) and row['order_date'] >= row['join_date'] and row['order_date']<= row['join_date'] +pd.Timedelta(days =6):
        return row['price']*10*2
  elif row['product_name'] == 'sushi':
         return row['price']*10*2
  else:
    return row['price']*10


sales_members['points'] = sales_members.apply(tot_points,axis=1)

membership_points =sales_members[sales_members['order_date'].dt.month==1].groupby('customer_id').agg(jan_points=('points','sum'))
membership_points
```
